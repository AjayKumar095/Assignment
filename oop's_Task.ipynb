```python
"""Q1, Create a vehicle class with an init method having instance variables as name_of_vehicle, max_speed
and average_of_vehicle."""

class Vehicle:
    
    def __init__(self, name_of_vehicle, max_speed, average_of_vehicle):
        
        self.name_of_vehicle=name_of_vehicle
        self.max_speed=max_speed
        self.average_of_vehicle=average_of_vehicle
        
    def vehicle_information(self):
        return f'Name: {self.name_of_vehicle} \nSpeed: {self.max_speed} \nAverage: {self.average_of_vehicle}'
    
    def vehicle_name(self):
        return f'Vehicle name: {self.name_of_vehicle}'

vehical=Vehicle('Honda', "240 Km/h", '13 Km/h')

print(f"Vechical Information :- \n\n{vehical.vehicle_information()}")      
```

```python
"""Q2. Create a child class car from the vehicle class created in Que 1, which will inherit the vehicle class.
Create a method named seating_capacity which takes capacity as an argument and returns the name of
the vehicle and its seating capacity."""

class Vehicle:
    
    def __init__(self, name_of_vehicle, max_speed, average_of_vehicle):
        
        self.name_of_vehicle=name_of_vehicle
        self.max_speed=max_speed
        self.average_of_vehicle=average_of_vehicle
        
    def vehicle_information(self):
        return f'Name: {self.name_of_vehicle} \nSpeed: {self.max_speed} \nAverage: {self.average_of_vehicle}'
    
    def vehicle_name(self):
        return f'Vehicle name: {self.name_of_vehicle}'

class Car(Vehicle):
    
    def __init__(self, name_of_vehicle, max_speed, average_of_vehicle, seating_capacity):
        super().__init__(name_of_vehicle, max_speed, average_of_vehicle)
        self.seating_capacity = seating_capacity
        
    def get_car_name(self):
        return f'Name of vehicle: {self.name_of_vehicle}\nSeating Capacity: {self.seating_capacity}'

# Creating an instance of the Car class
car = Car("Sedan", 200, 25, 5)

# Printing the name and seating capacity of the vehicle
print(f"Name and seating capacity of vehicle:\n{car.get_car_name()}")
  
```

```python
"""Q3. What is multiple inheritance? Write a python code to demonstrate multiple inheritance."""

class Animal:
    def __init__(self, species):
        self.species = species

    def display_species(self):
        print(f"Species: {self.species}")

class Mammal:
    def __init__(self, sound):
        self.sound = sound

    def make_sound(self):
        print(f"Sound: {self.sound}")

class Dog(Animal, Mammal):
    def __init__(self, name, breed, sound):
        # Calling the constructors of both parent classes
        Animal.__init__(self, species="Dog")
        Mammal.__init__(self, sound=sound)
        self.name = name
        self.breed = breed

    def display_info(self):
        print(f"Name: {self.name}\nBreed: {self.breed}")
        self.display_species()
        self.make_sound()

# Creating an instance of the Dog class
my_dog = Dog(name="Buddy", breed="Golden Retriever", sound="Woof!")

# Displaying information about the dog
my_dog.display_info()
```

```python
"""Q4. What are getter and setter in python? Create a class and create a getter and a setter method in this
class."""

class Student:
    def __init__(self, name, age):
        self._name = name  # Prefixing with an underscore to indicate it's a protected attribute
        self._age = age

    # Getter method for 'name'
    def get_name(self):
        return self._name

    # Setter method for 'name'
    def set_name(self, new_name):
        self._name = new_name

    # Getter method for 'age'
    def get_age(self):
        return self._age

    # Setter method for 'age'
    def set_age(self, new_age):
        if new_age > 0:
            self._age = new_age
        else:
            print("Age must be a positive value.")

# Creating an instance of the Student class
student1 = Student(name="John", age=20)

# Using getter methods
print(f"Original Name: {student1.get_name()}")
print(f"Original Age: {student1.get_age()}")

# Using setter methods
student1.set_name("Jane")
student1.set_age(22)

# Displaying updated information using getter methods
print(f"Updated Name: {student1.get_name()}")
print(f"Updated Age: {student1.get_age()}")
```

```python
"""Q5.What is method overriding in python? Write a python code to demonstrate method overriding."""

class Animal:
    def speak(self):
        print("Generic animal sound")

class Dog(Animal):
    # Overriding the speak method in the subclass
    def speak(self):
        print("Woof! Woof!")

class Cat(Animal):
    # Overriding the speak method in the subclass
    def speak(self):
        print("Meow!")

# Creating instances of the subclasses
dog = Dog()
cat = Cat()

# Calling the overridden speak method for each instance
dog.speak()  # Output: Woof! Woof!
cat.speak()  # Output: Meow!
```
