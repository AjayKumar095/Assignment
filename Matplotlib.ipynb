<h3>Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.</h3>

<p><b>Answer.</b></p>
<p>Matplotlib is a popular Python library used for creating static, interactive, and animated visualizations in Python. It provides a wide variety of plots and customization options, making it suitable for various data visualization needs.</p>
<p>It is used for visualizing data and presenting it in a graphical format, which aids in understanding patterns, trends, and relationships within the data. Matplotlib is widely used in fields such as data analysis, scientific computing, machine learning, and more.</p>
<p>Five plots that can be plotted using the Pyplot module of Matplotlib are:</p>
<ol>
  <li>Line plot</li>
  <li>Scatter plot</li>
  <li>Histogram</li>
  <li>Bar plot</li>
  <li>Pie chart</li>
</ol>

```python
"""Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot."""

import numpy as np 
import matplotlib.pyplot as plt 
np.random.seed(3)
x=3+np.random.normal(0,2,50)
y=3+np.random.normal(0,2,len(x))

plt.title("Scatter Plot")  # adding title
plt.xlabel("x-axis")   # adding x lable
plt.ylabel('y-axis')   # adding y lable
plt.scatter(x,y)  
plt.show()
```

```python
"""Q3: Why is the subplot() function used? Draw four line plots using the subplot() function."""

"""Answer
    The subplot() function in Matplotlib is used to create multiple plots within the same figure. It allows you to arrange plots in a grid of rows and columns, specifying the position of each subplot.
    This function is useful when you want to compare multiple plots or visualize different aspects of your data side by side."""
import numpy as np
import matplotlib.pyplot as plt

# Data for line plots
x = np.array([0, 1, 2, 3, 4, 5])

# Data for line 1
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for line 2
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for line 3
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for line 4
y4 = np.array([200, 350, 250, 550, 450, 150])

# Creating subplots
plt.figure(figsize=(10, 8))

# Subplot 1
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('Line 1')

# Subplot 2
plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('Line 2')

# Subplot 3
plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('Line 3')

# Subplot 4
plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('Line 4')

plt.tight_layout()
plt.show()
```

```python

"""Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot."""

"""Answer
        
A bar plot is a type of graph that represents categorical data with rectangular bars. The length of each bar corresponds to the value of the data it represents.
Bar plots are commonly used to compare the values of different categories or to show the distribution of categorical data."""

import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])


import numpy as np
import matplotlib.pyplot as plt

# Provided data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Creating a vertical bar plot
plt.figure(figsize=(8, 6))
plt.bar(company, profit, color='skyblue')
plt.xlabel('Company')
plt.ylabel('Profit (in billions)')
plt.title('Bar Plot of Company Profits')
plt.grid(True)
plt.show()

# Creating a horizontal bar plot
plt.figure(figsize=(8, 6))
plt.barh(company, profit, color='lightgreen')
plt.xlabel('Profit (in billions)')
plt.ylabel('Company')
plt.title('Horizontal Bar Plot of Company Profits')
plt.grid(True)
plt.show()
```

```python
"""Q5: What is a box plot? Why is it used? Using the following data plot a box plot."""

"""Answer
        Box plots are useful for visualizing the spread, skewness, and central tendency of the data, as well as identifying potential outliers. """

import matplotlib.pylab as plt 
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# creating box plot of given data
plt.boxplot([box1, box2])
plt.title('Boxplot')
plt.figure(figsize=(8,6))
plt.show()
```
